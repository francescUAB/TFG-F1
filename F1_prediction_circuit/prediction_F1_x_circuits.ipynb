```python

#llista de circuits 2024 i 2025
list_of_circuits=[1,3,4,6,7,9,11,13,14,15,17,18,21,22,24,39,70,73,77,78,79,80]


dictionary_circuits={
    1: 'albert_park',
    3: 'bahrain',
    4: 'catalunya',
    6: 'monaco',
    7: 'villeneuve',
    9: 'silverstone',
    11: 'hungaroring',
    13: 'spa',
    14: 'monza',
    15: 'marina_bay',
    17: 'shanghai',
    18: 'interlagos',
    21: 'imola',
    22: 'suzuka',
    80: 'vegas',
    24: 'yas_marina',
    39: 'zandvoort',
    70: 'red_bull_ring',
    73: 'baku',
    77: 'jeddah',
    78: 'losail',
    79: 'miami'
}
 

for circuit_id,circuit_name in dictionary_circuits.items():
        process_f1_all_positions_dataset(circuit_id)
```

```python
dir = './datasets'

archive_names = [archive for archive in os.listdir(dir) if os.path.isfile(os.path.join(dir, archive))]
dataframes_circuits=[]
null_counts = pd.DataFrame(columns=['File', 'Null Values'])

for archive in archive_names:
    file_path = os.path.join(dir, archive)
    if os.path.exists(file_path):
        dataframe = pd.read_csv(file_path, delimiter=';')
        dataframes_circuits.append(dataframe)
        null_count = dataframe.isnull().sum().sum()
        new_row = pd.DataFrame({'File': [file_path], 'Null Values': [null_count]})
        null_counts = pd.concat([null_counts, new_row], ignore_index=True)
    else:
        print(f"File {file_path} does not exist.")
        
print(null_counts)
```

```python
import matplotlib.pyplot as plt

results_summary = []  # Llista per emmagatzemar resultats resumits

for i, dataframe in enumerate(dataframes_circuits):
    
    df_filtered = dataframe[(dataframe['target'] <= 20) & (dataframe['year'] <= 2023)].dropna()


    if df_filtered.shape[0] == 0:
        print(f"No hay datos disponibles para Circuit_{i+1} tras el filtrado. Saltando este circuito.")
        continue

    df_train = df_filtered[df_filtered['year'] < 2021]
    df_valid = df_filtered[df_filtered['year'] == 2023]

   
    if df_train.shape[0] == 0 or df_valid.shape[0] == 0:
        print(f"No hay suficientes datos para Circuit_{i+1}. Saltando este circuito.")
        continue

    X_train = df_train.drop(columns=['dob', 'date', 'driverref', 'circuitref', 'target', 'raceid', 'positionorder', 'year'])
    y_train = df_train['target']

    X_test = df_valid.drop(columns=['dob', 'date', 'driverref', 'circuitref', 'target', 'raceid', 'positionorder', 'year'])
    y_test = df_valid['target']

    if len(y_train) < 30 or len(np.unique(y_train)) < 2 or min(y_train.value_counts()) < 6:
        print(f"No hay suficientes datos para Circuit_{i+1}. Saltando este circuito.")
        continue

    try:
        scaler = StandardScaler()
        X_train_scaled = scaler.fit_transform(X_train)
        X_test_scaled = scaler.transform(X_test)
    except ValueError as e:
        print(f"Error al escalar datos para Circuit_{i+1}: {e}")
        continue

    try:
        minority_class_count = min(y_train.value_counts())
        k_neighbors = max(1, min(5, minority_class_count - 1))  # Ajustar número de veïns
        smote = SMOTE(random_state=42, k_neighbors=k_neighbors)
        X_train_smote, y_train_smote = smote.fit_resample(X_train_scaled, y_train)
    except ValueError as e:
        print(f"Error al aplicar SMOTE para Circuit_{i+1}: {e}")
        continue

    rf = RandomForestClassifier(random_state=42)
    rf.fit(X_train_smote, y_train_smote)

    y_pred = rf.predict(X_test_scaled)
    accuracy = accuracy_score(y_test, y_pred)
    classification_rep = classification_report(y_test, y_pred, output_dict=True)
    f1 = classification_rep['weighted avg']['f1-score']
    recall = classification_rep['weighted avg']['recall']

    circuit_name = dictionary_circuits[list_of_circuits[i]]
    results_summary.append({
        'Circuit': circuit_name,
        'Accuracy': accuracy,
        'F1-Score': f1,
        'Recall': recall
    })
    print(f"\nResultados para Circuit_{circuit_name}:")
    print(f"Precisión del modelo: {accuracy:.4f}")
    print(f"F1-Score: {f1:.4f}")
    print(f"Recall: {recall:.4f}")

results_df = pd.DataFrame(results_summary)

print("\nResum de resultats per circuit:")
print(results_df)

results_df.to_csv('circuit_results_summary.csv', index=False)



plt.figure(figsize=(12, 6))
plt.bar(results_df['Circuit'], results_df['Accuracy'], color='skyblue', edgecolor='black')
plt.title('Accuracy per circuit - Random Forest', fontsize=20)
plt.xlabel('Circuit', fontsize=20)
plt.ylabel('Accuracy', fontsize=20)
plt.xticks(rotation=45, ha='right', fontsize=15)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from imblearn.over_sampling import SMOTE
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
from tensorflow.keras.optimizers import Adam
from sklearn.metrics import classification_report

results_summary = []

for i, dataframe in enumerate(dataframes_circuits):
    df_filtered = dataframe[(dataframe['target'] <= 20) & (dataframe['year'] <= 2023)].dropna()

    df_train = df_filtered[df_filtered['year'] < 2021]
    df_valid = df_filtered[df_filtered['year'] == 2023]

    if df_train.shape[0] == 0 or df_valid.shape[0] == 0:
        print(f"No hay suficientes datos para Circuit_{i+1}. Saltando este circuito.")
        continue


    X_train = df_train.drop(columns=['dob', 'date', 'driverref', 'circuitref','target', 'raceid', 'positionorder', 'year'])
    y_train = df_train['target']

    X_test = df_valid.drop(columns=['dob', 'date', 'driverref', 'circuitref', 'target', 'raceid', 'positionorder', 'year'])
    y_test = df_valid['target']

    scaler = StandardScaler()
    X_train_scaled = scaler.fit_transform(X_train)
    X_test_scaled = scaler.transform(X_test)

    min_samples_per_class = y_train.value_counts().min()
    if min_samples_per_class <= 1:
        print(f"Datos insuficientes para aplicar SMOTE en Circuit_{i+1}. Saltando este circuito.")
        continue
    k_neighbors = min(5, min_samples_per_class - 1)  # Mínimo de 2 muestras por clase para SMOTE
    smote = SMOTE(random_state=42, k_neighbors=k_neighbors)
    X_train_smote, y_train_smote = smote.fit_resample(X_train_scaled, y_train)

    encoder = OneHotEncoder(sparse_output=False)
    y_train_smote_onehot = encoder.fit_transform(y_train_smote.values.reshape(-1, 1))
    y_test_onehot = encoder.transform(y_test.values.reshape(-1, 1))

    model = Sequential([
        Dense(128, input_dim=X_train_smote.shape[1], activation='relu'),
        Dropout(0.3),
        Dense(64, activation='relu'),
        Dropout(0.3),
        Dense(y_train_smote_onehot.shape[1], activation='softmax')  # Una sortida per a cada classe
    ])

    model.compile(optimizer=Adam(learning_rate=0.001),
                  loss='categorical_crossentropy',
                  metrics=['accuracy'])

    history = model.fit(X_train_smote, y_train_smote_onehot,
                        validation_data=(X_test_scaled, y_test_onehot),
                        epochs=50,
                        batch_size=32,
                        verbose=0)

    y_pred_probs = model.predict(X_test_scaled)
    y_pred_classes = np.argmax(y_pred_probs, axis=1)
    y_test_classes = np.argmax(y_test_onehot, axis=1)

    accuracy = np.mean(y_pred_classes == y_test_classes)
    classification_rep = classification_report(y_test_classes, y_pred_classes, output_dict=True)
    f1 = classification_rep['weighted avg']['f1-score']
    recall = classification_rep['weighted avg']['recall']

    circuit_name = dictionary_circuits[list_of_circuits[i]]
    results_summary.append({
        'Circuit': circuit_name,
        'Accuracy': accuracy,
        'F1-Score': f1,
        'Recall': recall
    })

    print(f"\nResultados para Circuit {circuit_name}:")
    print(f"Accuracy: {accuracy:.4f}")
    print(f"F1-Score: {f1:.4f}")
    print(f"Recall: {recall:.4f}")

    plt.figure(figsize=(12, 5))

    plt.subplot(1, 2, 1)
    plt.plot(history.history['loss'], label='Entrenament')
    plt.plot(history.history['val_loss'], label='Validació')
    plt.title(f'Pèrdua - Circuit {circuit_name}')
    plt.xlabel('Èpoques')
    plt.ylabel('Pèrdua')
    plt.legend()

    plt.subplot(1, 2, 2)
    plt.plot(history.history['accuracy'], label='Entrenament')
    plt.plot(history.history['val_accuracy'], label='Validació')
    plt.title(f'Precisió - Circuit {circuit_name}')
    plt.xlabel('Èpoques')
    plt.ylabel('Precisió')
    plt.legend()

    plt.tight_layout()
    plt.show()

results_df = pd.DataFrame(results_summary)

print("\nResum de resultats per circuit:")
print(results_df)


results_df.to_csv('nn_circuit_results_summary.csv', index=False)
```
